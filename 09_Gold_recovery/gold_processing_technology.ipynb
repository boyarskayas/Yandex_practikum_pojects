#### Проверка расчета эффективности обогащения

Формула для расчета эффективности обогащения:

Recovery = (C * (F - T)) / (F * (C - T)) * 100, где:

- C — доля золота в концентрате после флотации/очистки - 'rougher.output.concentrate_au';
- F — доля золота в сырье/концентрате до флотации/очистки - 'rougher.input.feed_au';
- T — доля золота в отвальных хвостах после флотации/очистки - 'rougher.output.tail_au'.

'rougher.output.recovery' - 1-й целевой признак, на котором будем проверять правильность расчета эффективности обогащения.

```python
recovery_new = full['rougher.output.concentrate_au'] * (full['rougher.input.feed_au'] -
full['rougher.output.tail_au'])/(full['rougher.input.feed_au'] * (full['rougher.output.concentrate_au'] -
full['rougher.output.tail_au']))*100
```

```python
print('MAE =', mean_absolute_error(full['rougher.output.recovery'], recovery_new))
```

Средняя абсолютная ошибка близка к 0, поэтому делаем вывод, что 'rougher.output.recovery' в обучающей выборке рассчитана верно.

#### Корреляция признаков

#### Посмотрим, как меняется концентрации металлов (Au, Ag, Pb) на различных этапах очистки. 

```python
metal_au = full[["rougher.input.feed_au", "rougher.output.concentrate_au",
                 "primary_cleaner.output.concentrate_au", "final.output.concentrate_au"]]
metal_ag = full[["rougher.input.feed_ag", "rougher.output.concentrate_ag",
                 "primary_cleaner.output.concentrate_ag", "final.output.concentrate_ag"]]
metal_pb = full[["rougher.input.feed_pb", "rougher.output.concentrate_pb",
                 "primary_cleaner.output.concentrate_pb", "final.output.concentrate_pb"]]
```

```python
fig, axes = plt.subplots(3, 1, figsize=(18, 25))

for column in list(metal_au):
    sns.kdeplot(metal_au[column], ax=axes[0], shade = True, vertical=False, legend=True, cmap = "winter_r")
axes[0].set(title="Au", xlabel="Концентрация %", ylabel="Количество")

for column in list(metal_ag):
    sns.kdeplot(metal_ag[column], ax=axes[1], shade = True, vertical=False, legend=True, cmap = "winter_r")
axes[1].set(title="Ag", xlabel="Концентрация %", ylabel="Количество")

for column in list(metal_pb):
    sns.kdeplot(metal_pb[column], ax=axes[2], shade = True, vertical=False, legend=True, cmap = "winter_r")
axes[2].set(title="Pb", xlabel="Концентрация %", ylabel="Количество")

fig.suptitle("Концентрация металлов на стадиях очистки")
fig.legend(["rougher.input.feed", "rougher.output.concentrate", "primary.cleaner.output concentrate", "final.output.concentrate",])
fig.show()
```

ВЫВОДЫ:  

На всех графиках распределение близко к нормальному. У всех металлов наблюдаем маленький пик около 0. Это говорит о том, что присутствует концентрация металлов на всех этапах.  
- Концентрация серебра увеличивается на этапе флотации, а затем постепенно уменьшается.
- Концентрация золота после каждого этапа очитски постепенно увеличивается. Максимальная концентрация на финальном этапе обработки, пик приходится на ≈45.
- Концентрация свинца сначала увеличивается на этапе флотации и после первичной очитски, затем остатется примерно на одном уровне.

#### Функции для вычисления sMape

Напишем функцию для вычисления sMape. Для избежания дублирования кода сначала используем функцию для невзвешенной sMAPE. 

```python
def get_smape(target, predict):
    smape = (1/len(target))*np.sum((np.abs(target - predict))/( (np.abs(target)+np.abs(predict))/2))*100
    return smape
```

```python
smape_scorer = make_scorer(get_smape, greater_is_better = False)
```

Напишем функцию для вычисления итоговой sMAPE.

```python
def get_final_smape(rougher, final):
    final = 0.25*rougher+0.75*final
    return final
```

#### Обучим разные модели и оценим их качество кросс-валидацией

**Модель LinearRegression**

```python
%%time
model_lr_rougher = LinearRegression()
smape_rougher = cross_val_score(model_lr_rougher,
                          features_train,
                          target_rougher,
                          scoring=smape_scorer,
                          cv=5)
final_smape_rougher = smape_rougher.mean()

model_lr_final = LinearRegression()
smape_final = cross_val_score(model_lr_final,
                        features_train,
                        target_final,
                        scoring=smape_scorer,
                        cv=5)
final_smape_final = smape_final.mean()

result_smape = get_final_smape(final_smape_rougher, final_smape_final)*(-1)

scores_dct['LR'] = [result_smape]
```

**Модель DecisionTreeRegressor**

```python
%%time
best_result_for_dtr = 100
best_param_for_dtr = 0

for n in range(1, 10, 1):
    model_dtr_rougher = DecisionTreeRegressor(max_depth=n, random_state=state)
    smape_rougher = cross_val_score(model_dtr_rougher,
                              features_train,
                              target_rougher,
                              scoring=smape_scorer,
                              cv=5)
    final_smape_rougher = smape_rougher.mean()

    model_dtr_final = DecisionTreeRegressor(max_depth=n, random_state=state)
    smape_final = cross_val_score(model_dtr_final, 
                                  features_train,
                                  target_final,
                                  scoring=smape_scorer,
                                  cv=5)
    final_smape_final = smape_final.mean()
    
    result_smape = get_final_smape(final_smape_rougher, final_smape_final)*(-1)
    
    if result_smape < best_result_for_dtr:
        best_result_for_dtr = result_smape
        best_param_for_dtr = n
        
scores_dct['DTR'] = [best_result_for_dtr, best_param_for_dtr]
```

**Модель RandomForestRegressor**

```python
%%time
best_model_for_rfr = 100
best_param_for_rfr = 0

for n in range(10, 100, 30):
    for d in range(2,10,2):
        model_rfr_rougher = RandomForestRegressor(n_estimators=n, max_depth=d, random_state=state)
        smape_rougher = cross_val_score(model_rfr_rougher,
                                  features_train,
                                  target_rougher,
                                  scoring=smape_scorer,
                                  cv=5)
        final_smape_rougher = smape_rougher.mean()

        model_rfr_final = RandomForestRegressor(n_estimators=n, max_depth=d, random_state=state)
        smape_final = cross_val_score(model_rfr_final,
                                features_train,
                                target_final,
                                scoring=smape_scorer,
                                cv=5)
        final_smape_final = smape_final.mean()
        
        result_smape = get_final_smape(final_smape_rougher, final_smape_final)*(-1)
        
        if result_smape < best_model_for_rfr:
            best_model_for_rfr = result_smape
            best_param_for_rfr = [n,d] 
            
scores_dct['RFR'] = [best_model_for_rfr, best_param_for_rfr]
```
