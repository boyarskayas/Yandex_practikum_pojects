#### Скользящее среднее

*Скользящее среднее (англ. rolling mean), или движущееся среднее (англ. moving average), — метод сглаживания временных рядов. Его суть заключается в поиске значений, которые меньше всего подвержены колебаниям, то есть средних арифметических. Метод работает так: экспериментально подбирается размер окна (англ. window size) — интервал, в котором выполняют усреднение. Чем интервал больше, тем сильнее сглаживание, то есть суть заключается в поиске значений, которые меньше всего подвержены колебаниям (средние арифметические).*

```python
def roll(df, rolling):
    df['rolling_mean'] = df.rolling(rolling).mean()
    df['std'] = df['num_orders'].rolling(rolling).std()
    df = df.dropna()
    df.plot(figsize=(15,7))
    plt.title('Заказы такси со скользящим средним')
    plt.legend()
    plt.show()
```

```python
roll(df, 24)
```

```python
# чем выше размер скользящего окна, тем сильнее сглаживание
roll(df, 168)
```

Среднее значение и стандартное отклонение со временем увеличиваются, значит, временной ряд нестационарный.

## Обучение

### Необходимые функции

Напишем функцию, добавлящую новые календарные признаки, а так же признаки сезонности, тренда и скользящего окна.  
Функция подает на выход данные, разделенные на трейн и таргет

```python
def make_features(df, max_lag, rolling_mean_size, test_size):
    # признаки по календарю: год, месяц, день, день недели, час
    df_new = df.copy()
    df_new['day'] = df_new.index.day
    df_new['dayofweek'] = df_new.index.dayofweek
    df_new['hour'] = df_new.index.hour
    # зададим отстающие значения
    for lag in range(1, max_lag + 1):
        df_new['lag_{}'.format(lag)] = df_new['num_orders'].shift(lag)
        
    # изменим столбец rolling_mean
    df_new['rolling_mean'] = df_new['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    # разобьем выборки на трейн и тест
    train, test = train_test_split(df_new, shuffle=False, test_size=test_size)
    train = train.dropna()
    
    # разобьем трейн и тест на фичи и таргеты
    train_target = train['num_orders']
    train_features = train.drop(['num_orders'], axis=1)
    test_features = test.drop('num_orders', axis=1)
    test_target = test[['num_orders']]
    
    return train_features, train_target, test_features, test_target
```

```python
# Получим выборки
train_features, train_target, test_features, test_target = make_features(df, 168, 168, 0.1)

train_features.shape, train_target.shape, test_features.shape, test_target.shape
```

Создадим словари гиперпараметров для обучения моделей и ансамблей

```python
dt_params = {
    'max_depth' : range(1,15)
}

rf_params = {
    'n_estimators' : range(50, 300, 50), 
    'max_depth' : range(5, 15, 5)
}

lgbm_params = { 
    'num_leaves' : range(10,60,10),
    'learning_rate' : [0.2, 0.5]
}

cbr_params = {
    'learning_rate' : [0.1, 0,3, 0.5],
    'depth' : range(1, 16, 5)
#     'n_estimators' : range(100, 501, 100)
}
```

Напишем функцию для поиска лучшего RMSE каждой модели и ансамбля, используя RandomSearch 

```python
def Randomiz_search(model, params, num_iter, verb, cv_mod, X, y):
    model = RandomizedSearchCV(
        model,
        param_distributions = params,
        n_iter = num_iter,
        verbose = verb,
        scoring = 'neg_root_mean_squared_error',
        cv = TimeSeriesSplit(n_splits=cv_mod)
    )
    model.fit(X, y)
    RMSE = round((-model.best_score_), 2)
     
    return model, RMSE
```

### RMSE  линейной регрессии

```python
lin = []
```

```python
linear, linear_RMSE = Randomiz_search(
        LinearRegression(), 
        {}, 5, 200, 5,  
        train_features, train_target
)

print('RMSE линейной регресии равен:', linear_RMSE)

lin.append(linear_RMSE)
```

### RMSE  дерева решений

```python
dt_list = []
```

```python
dt, dt_RMSE = Randomiz_search(
    DecisionTreeRegressor(random_state=RND), 
    dt_params, 8, 0, 5, 
    train_features, train_target
)

print('RMSE решающего дерева равен:', dt_RMSE)

dt_list.append(dt_RMSE)
```

### RMSE случайного леса

```python
rf_list = []
```

```python
rf, rf_RMSE = Randomiz_search(
                RandomForestRegressor(random_state=RND), 
                rf_params, 
                5, 200, 5,
                train_features, train_target
)

print('RMSE случайного леса равен:', rf_RMSE)

rf_list.append(rf_RMSE)
```

### RMSE  LGBMRegressor

```python
light = []
```

```python
lgbm, lgbm_RMSE = Randomiz_search(
                    LGBMRegressor(
                        class_weight = 'balanced', 
                        random_state = RND
                                                    ), 
                    lgbm_params, 
                    5, 200, 5, 
                    train_features, train_target
)
print('RMSE LGBMRegressor равен:', lgbm_RMSE)

light.append(lgbm_RMSE)
```

### RMSE  CatBoostRegressor

```python
cat = []
```

```python
cbr, cbr_RMSE = Randomiz_search(
        CatBoostRegressor(
            grow_policy='Depthwise', 
            n_estimators=100,  
            random_state = RND
        ),
        cbr_params, 
        5, 200, 5, 
        train_features, train_target
)
print('RMSE CatBoostRegressor равен:', cbr_RMSE)

cat.append(cbr_RMSE)
```

### Сравним показатели RMSE разных моделей и ансамблей

```python
res_values = []
res_values.append(lin)
res_values.append(dt_list)
res_values.append(rf_list)
res_values.append(light)
res_values.append(cat)
```

```python
res = pd.DataFrame(res_values, 
                   columns = ['RMSE'],
                   index=['linear_regression', 
                          'DesicionTree', 
                          'RandomForest', 
                          'LGBMRegressor',
                          'CatBoostRegressor_RMSE'
                         ]
)
```
